```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
 

```

```python
df= pd.read_csv("wisc_bc_data.csv")
```

```python
describe_df=df.describe()

col=describe_df.columns[:]
col
```

```python
describe_df
```

```python
for i in col:
    
    from matplotlib import mlab

    mu,sigma,max_,min_,q1,q2,q3=describe_df.loc[['mean','std','max','min','25%','50%','75%'],i]
    
 
    count, bins, ignored = plt.hist(df[i], bins=45, density=False,color='g')

    #curve=(1/(sigma * np.sqrt(2 * np.pi))) *( np.exp( - (bins - mu)**2 / (2 * sigma**2) ))
    
    #y = mlab.normpdf(bins, mu, sigma)
    #l = plt.plot(bins,  600*y, 'r--', linewidth=1)

    plt.axvline(x=mu,linewidth=4, color='y')
    plt.axvline(x=mu-sigma,linewidth=2, color='r')
    plt.axvline(x=mu+sigma,linewidth=2, color='r')
    plt.axvline(x=q1,linewidth=3, color='m')
    plt.axvline(x=q2,linewidth=3, color='m')
    plt.axvline(x=q3,linewidth=3, color='m')


    plt.text(mu,-6,'Mean',rotation=90,fontsize=10)
    plt.text(mu-sigma,-5,'-Sigma',rotation=90,fontsize=10)
    plt.text(mu+sigma,-5,'+Sigma',rotation=90,fontsize=10)
    plt.text(q1,-5,'Q1',rotation=90,fontsize=10)
    plt.text(q2,-5,'Q2',rotation=90,fontsize=10)
    plt.text(q3,-5,'Q3',rotation=90,fontsize=10)



    #plt.axvline(x=max_,linewidth=2, color='r')
    #plt.axvline(x=min_,linewidth=2, color='r')



    fig = plt.gcf()
    fig.set_size_inches(12.5, 8, forward=True)



    plt.show()
```
